## Data preprocessing

```python
# am using standard scaler to fit model 
StandardScaler=StandardScaler()
# am not applying columns_to_scale to dummies 
columns_to_scale = ['age','trestbps','chol','thalach','oldpeak']
df[columns_to_scale] = StandardScaler.fit_transform(df[columns_to_scale])
```

```python
dataset=pd.get_dummies(df,columns = ['sex','cp','fbs','restecg','exang','slope','ca','thal'])
dataset.head()
```

```python
y=dataset["target"] #  creating dependent feature
x=dataset.drop(["target"], axis =1) # indipendent feature
```

```python
# cross validation
knn_scores = []
for k in range(1,21): #  for iteration     am targeting 20 nearest nighbour
    knn_classifier = KNeighborsClassifier(n_neighbors =k)
    score=cross_val_score(knn_classifier,x,y,cv=10)# 10 is for 10 experiment values
    knn_scores.append(score.mean())
```

```python
plt.plot([k for k in range(1,21)], knn_scores,color ="red")
for i in range (1,21):
    plt.text(i, knn_scores[i-1],(i,knn_scores[i-1]))
plt.xticks([i for i in range(1,21)])
plt.xlabel("number of neighbors(k)")
plt.ylabel("scores")
plt.title("k Neighbors Classifier scores for different k values")
```
